```python
import random
import uuid
import statistics
import matplotlib.pyplot as plt
import json
import os

# Classe base
class Pessoa:
    def __init__(self, nome, email):
        self.nome = nome
        self.email = email
        self.id = str(uuid.uuid4())

    def exibir_dados(self):
        return f"id:{self.id}: {self.nome}\n Email: {self.email}\n"

# Classe Aluno
class Aluno(Pessoa):
    def __init__(self, nome, email, matricula):
        super().__init__(nome, email)
        self.matricula = matricula

    def exibir_dados(self):
        return f"[Aluno] {super().exibir_dados()} | Matr√≠cula: {self.matricula}"

# Classe Professor
class Professor(Pessoa):
    def __init__(self, nome, email, disciplina):
        super().__init__(nome, email)
        self.disciplina = disciplina

    def exibir_dados(self):
        return f"[Professor] {super().exibir_dados()} \n Disciplina: {self.disciplina}"

# Sistema de Cadastro
class SistemaCadastro:
    def __init__(self):
        self.alunos = []
        self.professores = []
        self.notas_quiz = []
        self.carregar_dados()

    def salvar_dados(self):
        with open("alunos.json", "w") as f:
            json.dump([vars(a) for a in self.alunos], f)
        with open("professores.json", "w") as f:
            json.dump([vars(p) for p in self.professores], f)
        with open("notas.json", "w") as f:
            json.dump(self.notas_quiz, f)

    def carregar_dados(self):
        if os.path.exists("alunos.json"):
            with open("alunos.json", "r") as f:
                alunos_data = json.load(f)
                self.alunos = [Aluno(d["nome"], d["email"], d["matricula"]) for d in alunos_data]

        if os.path.exists("professores.json"):
            with open("professores.json", "r") as f:
                professores_data = json.load(f)
                self.professores = [Professor(d["nome"], d["email"], d["disciplina"]) for d in professores_data]

        if os.path.exists("notas.json"):
            with open("notas.json", "r") as f:
                self.notas_quiz = json.load(f)

    def cadastrar_aluno(self):
        nome = input("Nome do aluno: ")
        email = input("Email do aluno: ")
        matricula = input("Matr√≠cula do aluno: ")
        aluno = Aluno(nome, email, matricula)
        self.alunos.append(aluno)
        self.salvar_dados()
        print(" Aluno cadastrado com sucesso!")

    def cadastrar_professor(self):
        nome = input("Nome do professor: ")
        email = input("Email do professor: ")
        disciplina = input("Disciplina do professor: ")
        professor = Professor(nome, email, disciplina)
        self.professores.append(professor)
        self.salvar_dados()
        print(" Professor cadastrado com sucesso!")

    def listar_todos(self):
        print("\n Lista de Alunos:")
        for aluno in self.alunos:
            print(aluno.exibir_dados())

        print("\n Lista de Professores:")
        for professor in self.professores:
            print(professor.exibir_dados())

    def listar_notas(self):
        if not self.notas_quiz:
            print("Nenhuma nota registrada ainda.")
            return

        print("\n Hist√≥rico de Notas dos Quizzes:")
        for nota in self.notas_quiz:
            print(f"Aluno: {nota['aluno']} | Nota: {nota['pontuacao']}/10")

    def calcular_estatisticas(self):
        if not self.notas_quiz:
            print("Nenhuma nota registrada para estat√≠sticas.")
            return

        notas = [nota['pontuacao'] for nota in self.notas_quiz]
        media = statistics.mean(notas)
        mediana = statistics.median(notas)
        try:
            moda = statistics.mode(notas)
        except statistics.StatisticsError:
            moda = "N√£o h√° uma moda (valores igualmente frequentes)"

        print("\nüìä Estat√≠sticas das Notas:")
        print(f"M√©dia: {media:.2f}")
        print(f"Mediana: {mediana}")
        print(f"Moda: {moda}")

        # Gr√°fico
        plt.figure(figsize=(8, 6))
        plt.hist(notas, bins=range(0, 12), edgecolor='black', alpha=0.7)
        plt.axvline(media, color='blue', linestyle='dashed', linewidth=2, label=f"M√©dia: {media:.2f}")
        plt.axvline(mediana, color='orange', linestyle='dashed', linewidth=2, label=f"Mediana: {mediana}")
        if isinstance(moda, (int, float)):
            plt.axvline(moda, color='green', linestyle='dashed', linewidth=2, label=f"Moda: {moda}")
        plt.title('Distribui√ß√£o das Notas do Quiz')
        plt.xlabel('Notas')
        plt.ylabel('Frequ√™ncia')
        plt.legend()
        plt.grid(True)
        plt.savefig("grafico_notas.png")
        plt.show(block=False)

# Quiz
class Quiz:
    def __init__(self, aluno_nome, sistema_cadastro):
        self.aluno_nome = aluno_nome
        self.sistema_cadastro = sistema_cadastro
        self.pontuacao = 0
        self.perguntas = [
            {"pergunta": "\n Qual estrutura √© usada para repetir um bloco de c√≥digo um n√∫mero fixo de vezes? \n", "opcoes": ["if", "for", "while", "elif"], "correta": "for"},
            {"pergunta": "\n Qual estrutura repete enquanto uma condi√ß√£o for verdadeira? \n", "opcoes": ["for", "if", "while", "else"], "correta": "while"},
            {"pergunta": "\n Qual palavra √© usada para criar uma condi√ß√£o em Python? \n", "opcoes": ["loop", "for", "if", "repeat"], "correta": "if"},
            {"pergunta": "\n O que ser√° impresso?\nfor i in range(3): print(i) \n", "opcoes": ["1 2 3", "0 1 2", "0 1 2 3", "1 2"], "correta": "0 1 2"},
            {"pergunta": "\n Qual dessas op√ß√µes √© usada para tratar uma segunda condi√ß√£o? \n", "opcoes": ["elif", "else if", "otherwise", "while"], "correta": "elif"},
            {"pergunta": "\n Quando usamos 'else'? \n", "opcoes": ["Quando a condi√ß√£o √© verdadeira", "Sempre que 'if' falha", "Quando queremos repetir algo", "Para encerrar o programa"], "correta": "Sempre que 'if' falha"},
            {"pergunta": "\n Qual √© o valor final de 'x'?\nx = 0\nfor i in range(3):\n    x += 1 \n", "opcoes": ["2", "3", "0", "1"], "correta": "3"},
            {"pergunta": "\n Qual destas estruturas pode causar um loop infinito? \n", "opcoes": ["for", "while", "if", "elif"], "correta": "while"},
            {"pergunta": "\n O que esse c√≥digo faz?\nif 5 > 3:\n    print('A')\nelse:\n    print('B')\n", "opcoes": ["Imprime B", "Erro", "Nada", "Imprime A"], "correta": "Imprime A"},
            {"pergunta": "\n Como encerrar um loop antecipadamente? \n", "opcoes": ["continue", "end", "break", "stop"], "correta": "break"}
        ]

    def iniciar(self):
        print("\n Iniciando o Quiz sobre Repeti√ß√£o e Condicionais!\n")
        random.shuffle(self.perguntas)

        explicacoes = {
            "for": " O 'for' √© usado quando sabemos exatamente quantas vezes queremos repetir um bloco de c√≥digo.",
            "while": " O 'while' repete enquanto a condi√ß√£o for verdadeira. Cuidado com loops infinitos!",
            "if": " O 'if' √© usado para testar condi√ß√µes e decidir o fluxo do programa.",
            "0 1 2": " A fun√ß√£o range(3) gera os valores 0, 1 e 2, por isso o print exibe 0 1 2.",
            "elif": " O 'elif' permite testar outra condi√ß√£o se o 'if' inicial falhar.",
            "Sempre que 'if' falha": " O 'else' √© executado quando a condi√ß√£o do 'if' n√£o √© satisfeita.",
            "3": " O loop for √© executado 3 vezes, e a vari√°vel x √© incrementada em cada uma, ficando com valor 3.",
            "Imprime A": " O c√≥digo verifica se 5 > 3, o que √© verdadeiro, ent√£o imprime 'A'.",
            "break": " O 'break' encerra um loop imediatamente, √∫til para sair antes do final.",
        }

        for i, q in enumerate(self.perguntas, 1):
            print(f"\nüîπ Quest√£o {i}")
            print(q["pergunta"])
            for idx, opcao in enumerate(q["opcoes"], 1):
                print(f"{idx}. {opcao}")
            resposta = input("Sua resposta (n√∫mero): ")

            if resposta.isdigit():
                indice = int(resposta) - 1
                if 0 <= indice < len(q["opcoes"]):
                    if q["opcoes"][indice] == q["correta"]:
                        print(" Correto!")
                        self.pontuacao += 1
                    else:
                        print(f" Errado! Resposta correta: {q['correta']}")
                    print(explicacoes.get(q["correta"], " Sem explica√ß√£o dispon√≠vel."))
                else:
                    print(" Op√ß√£o fora do alcance.")
            else:
                print(" Resposta inv√°lida.")

        print(f"\n Fim do quiz! Sua pontua√ß√£o: {self.pontuacao}/10")
        self.sistema_cadastro.notas_quiz.append({
            "aluno": self.aluno_nome,
            "pontuacao": self.pontuacao
        })
        self.sistema_cadastro.salvar_dados()

# Login
def autenticar_usuario(sistema):
    print("\n[ Login ]")
    tipo = input("Voc√™ √© (1) Aluno ou (2) Professor? ")
    nome = input("Digite seu nome: ")

    if tipo == "1":
        for aluno in sistema.alunos:
            if aluno.nome == nome:
                return "aluno", aluno
        print("Aluno n√£o encontrado.")
    elif tipo == "2":
        for professor in sistema.professores:
            if professor.nome == nome:
                return "professor", professor
        print("Professor n√£o encontrado.")
    else:
        print("Op√ß√£o inv√°lida.")
    return None, None

# Menu do Aluno
def menu_aluno(aluno, sistema):
    while True:
        print(f"\nBem-vindo, {aluno.nome} [Aluno]")
        print("1. Iniciar o quiz")
        print("2. Ver hist√≥rico de notas")
        print("3. Ver estat√≠sticas e gr√°fico")
        print("0. Voltar")
        escolha = input("Op√ß√£o: ")

        if escolha == "1":
            quiz = Quiz(aluno.nome, sistema)
            quiz.iniciar()
        elif escolha == "2":
            sistema.listar_notas()
        elif escolha == "3":
            sistema.calcular_estatisticas()
        elif escolha == "0":
            break
        else:
            print("‚ö†Ô∏è Op√ß√£o inv√°lida.")

# Menu do Professor
def menu_professor(professor, sistema):
    while True:
        print(f"\nBem-vindo, {professor.nome} [Professor]")
        print("1. Ver todos os usu√°rios")
        print("2. Ver hist√≥rico de notas")
        print("3. Ver estat√≠sticas e gr√°fico")
        print("0. Voltar")
        escolha = input("Op√ß√£o: ")

        if escolha == "1":
            sistema.listar_todos()
        elif escolha == "2":
            sistema.listar_notas()
        elif escolha == "3":
            sistema.calcular_estatisticas()
        elif escolha == "0":
            break
        else:
            print("‚ö†Ô∏è Op√ß√£o inv√°lida.")

# Menu principal
def menu():
    sistema = SistemaCadastro()

    while True:
        print("\n========= MENU =========")
        print("1. Login como Aluno")
        print("2. Login como Professor")
        print("3. Cadastrar Aluno")
        print("4. Cadastrar Professor")
        print("0. Sair")
        print("========================")
        opcao = input("Escolha uma op√ß√£o: ")

        if opcao == "1":
            tipo, usuario = autenticar_usuario(sistema)
            if tipo == "aluno":
                menu_aluno(usuario, sistema)
        elif opcao == "2":
            tipo, usuario = autenticar_usuario(sistema)
            if tipo == "professor":
                menu_professor(usuario, sistema)
        elif opcao == "3":
            sistema.cadastrar_aluno()
        elif opcao == "4":
            sistema.cadastrar_professor()
        elif opcao == "0":
            print("Encerrando o sistema...")
            break
        else:
            print("‚ö†Ô∏è Op√ß√£o inv√°lida.")

if __name__ == "__main__":
    menu()

```
